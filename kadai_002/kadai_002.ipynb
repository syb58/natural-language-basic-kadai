```python
import re
from bs4 import BeautifulSoup
from urllib import request
import MeCab
import unidic_lite

#メインテキストの抽出
url = "https://www.aozora.gr.jp/cards/000148/files/2371_13943.html"
response = request.urlopen(url)
soup = BeautifulSoup(response)
response.close()

main_text = soup.find('div', class_='main_text')
# tags_to_delete = main_text.find_all(['rp', 'rt', 'rb' ,'br', 'ruby', 'strong'])
# for tag in tags_to_delete:
#   tag.decompose()

# ストップワードの抽出
main_text = main_text.get_text()
main_text = re.sub(r"[\u3000 \n \r]", "", main_text)
print(main_text)

stop_word_url = 'http://svn.sourceforge.jp/svnroot/slothlib/CSharp/Version1/SlothLib/NLP/Filter/StopWord/word/Japanese.txt'
response_word = request.urlopen(stop_word_url)
soup_word =BeautifulSoup(response_word)
response_word.close()
soup_word = soup_word.text

stopword_list = soup_word.split("\r\n")
stopword_list = [word for word in stopword_list if word]
print(stopword_list)

# メインテキストの形態素解析、単語のみ分かち文でリスト作成
mecab_tagger = MeCab.Tagger('-Owakati')
text = main_text
result_text = mecab_tagger.parse(main_text).split()
print(result_text)

# メインテキストとストップワードを参照。ストップワードの除去
result_text_list = list()
for split_text in result_text:
  if split_text not in stopword_list:
    result_text_list.append(split_text)

print(result_text_list)
```
